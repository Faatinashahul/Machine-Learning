## DESIGN

```python
# DESIGN IMPLEMENTTAION- MATHEMATICAL APPROACH

print("")
print("FORMULA 1: Euclidean Distance")
print("    d = sqrt root of[summation(xi - ci)^2]")
print("    where:")
print("    • xi = ith feature of input")
print("    • ci = ith feature of center (target)")
print("")

print("FORMULA 2: Gaussian Activation")
print("    similarity = exp(-d^2 / 2(variance)²)")
print("    where:")
print("    • d = distance from Formula 1")
print("    • variance = spread parameter")
print("")
print("COMPLETE NEURON FUNCTION:")
print("    Output = exp(-[Σ(xᵢ - cᵢ)²] / 2σ²)")
print("")

print("PROPERTIES:")
print("    • When x = c → d = 0 → Output = 1 (Maximum)")
print("    • When x → ∞ → d → ∞ → Output → 0 (Minimum)")
print("    • σ controls how quickly output drops with distance")
```

## IMPLEMENTATION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns


class CustomRBFNeuron:
    def __init__(self, center, sigma=1.0, class_label=None):
        self.center = np.array(center)
        self.sigma = sigma
        self.class_label = class_label 
    
    def activate(self, input_vector):
        input_vector = np.array(input_vector)
        distance = np.linalg.norm(input_vector - self.center)
        if self.sigma == 0:
            return 1.0 if distance == 0 else 0.0
        else:
            similarity = np.exp(- (distance**2) / (2 * (self.sigma**2)))
            return similarity
    
    def __repr__(self):
        return f"RBFNeuron(class={self.class_label}, center={self.center}, σ={self.sigma})"
```

```python
# Loading iris dataset
iris = load_iris()
X = iris.data  
y = iris.target  
feature_names = iris.feature_names
target_names = iris.target_names

print(f"Dataset shape: {X.shape}")
print(f"Number of samples: {len(X)}")
print(f"Number of features: {X.shape[1]}")
print(f"Features: {feature_names}")
print(f"Classes: {target_names}")
print(f"Class distribution: {np.bincount(y)}")
```

```python
print("\nFirst 5 samples:")
df = pd.DataFrame(X, columns=feature_names)
df['species'] = [target_names[i] for i in y]
df.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
print("\n")
print(f"Training class distribution: {np.bincount(y_train)}")
print(f"Testing class distribution: {np.bincount(y_test)}")


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n")
print("Features standardized (mean=0, std=1)")
print(f"Training data mean after scaling: {X_train_scaled.mean(axis=0).round(2)}")
print(f"Training data std after scaling: {X_train_scaled.std(axis=0).round(2)}")
```

```python
#using means as centers 
centers = []
neurons = []

for class_id in range(3):
    class_samples = X_train_scaled[y_train == class_id]
    center = np.mean(class_samples, axis=0)
    centers.append(center)
    distances = []
    for sample in class_samples:
        dist = np.linalg.norm(sample - center)
        distances.append(dist)
        sigma = np.mean(distances)  
        neuron = CustomRBFNeuron(center=center, sigma=sigma, class_label=target_names[class_id])
        neurons.append(neuron)

    print(f"\nNeuron for {target_names[class_id]}:")
    print(f"  Center: {center.round(3)}")
    print(f"  Spread (σ): {sigma:.3f}")
    print(f"  Number of training samples: {len(class_samples)}")
```

```python
print("TESTING NEURONS ON TRAINING SAMPLES")
def classify_sample(sample, neurons):
    outputs = []
    for neuron in neurons:
        similarity = neuron.activate(sample)
        outputs.append((neuron.class_label, similarity))
    outputs.sort(key=lambda x: x[1], reverse=True)
    return outputs[0]  

print("\nFirst 5 training samples:")
print("-" * 80)
print(f"{'True Class':<15} {'Predicted':<15} {'Similarity':<12} {'Confidence'}") 
print("-" * 80)
for i in range(5):
    sample = X_train_scaled[i]
    true_class = target_names[y_train[i]]
    predicted_class, similarity = classify_sample(sample, neurons)
    all_outputs = [(n.class_label, n.activate(sample)) for n in neurons]
    all_outputs.sort(key=lambda x: x[1], reverse=True)
    confidence = all_outputs[0][1] - all_outputs[1][1]
    print(f"{true_class:<15} {predicted_class:<15} {similarity:<12.4f} {confidence:.4f}")
```

```python
predictions = []
true_labels = []

for i in range(len(X_test_scaled)):
    sample = X_test_scaled[i]
    true_class = target_names[y_test[i]]
    predicted_class, similarity = classify_sample(sample, neurons)
    predictions.append(predicted_class)
    true_labels.append(true_class)
accuracy = accuracy_score(true_labels, predictions)
print(f"\nTest Accuracy: {accuracy:.2%} ({np.sum(np.array(true_labels) == np.array(predictions))}/{len(y_test)} correct)")


# Confusion Matrix
cm = confusion_matrix(true_labels, predictions, labels=target_names)
print("\nConfusion Matrix:")
print("-" * 40)
print("True\\Pred  " + "  ".join([f"{name[:4]:4}" for name in target_names]))
for i, true_name in enumerate(target_names):
    row = f"{true_name:<10}"
    for j in range(3):
        row += f" {cm[i, j]:4d}   "
    print(row)

# Per-class performance
print("\nPer-Class Performance:")
print("-" * 50)
print(f"{'Class':<15} {'Precision':<12} {'Recall':<12} {'Support'}")
print("-" * 50)

for i, class_name in enumerate(target_names):
    true_pos = cm[i, i]
    false_pos = sum(cm[:, i]) - true_pos
    false_neg = sum(cm[i, :]) - true_pos
    precision = true_pos / (true_pos + false_pos) if (true_pos + false_pos) > 0 else 0
    recall = true_pos / (true_pos + false_neg) if (true_pos + false_neg) > 0 else 0
    support = sum(y_test == i)
    print(f"{class_name:<15} {precision:<12.2f} {recall:<12.2f} {support}")



sample_indices = [0, 10, 20]  
sample_names = ['Sample A', 'Sample B', 'Sample C']
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for idx, (sample_idx, ax) in enumerate(zip(sample_indices, axes)):
    sample = X_test_scaled[sample_idx]
    true_class = target_names[y_test[sample_idx]]
    
    activations = []
    neuron_names = []
    for neuron in neurons:
        act = neuron.activate(sample)
        activations.append(act)
        neuron_names.append(neuron.class_label)
    
    bars = ax.bar(neuron_names, activations, color=['red', 'green', 'blue'])
    ax.set_ylim(0, 1)
    ax.set_ylabel('Activation (Similarity)')
    ax.set_title(f'Sample {idx+1}\nTrue Class: {true_class}')
    ax.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5, label='Decision Threshold')
    
    for bar, act in zip(bars, activations):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                f'{act:.2f}', ha='center', va='bottom')
plt.tight_layout()
plt.show()


X_2d = X_train_scaled[:, :2]
feature_names_2d = feature_names[:2]

x_min, x_max = X_2d[:, 0].min() - 0.5, X_2d[:, 0].max() + 0.5
y_min, y_max = X_2d[:, 1].min() - 0.5, X_2d[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

Z = np.zeros(xx.shape)
for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        point = np.array([xx[i, j], yy[i, j]])
        # We need all 4 features, but we only have 2 - use mean for missing features
        full_point = np.array([xx[i, j], yy[i, j], 
                               np.mean(X_train_scaled[:, 2]), 
                               np.mean(X_train_scaled[:, 3])])
        pred_class, _ = classify_sample(full_point, neurons)
        Z[i, j] = list(target_names).index(pred_class)

plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis', levels=3)

scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y_train, 
                      cmap='viridis', edgecolors='black', s=100)
plt.colorbar(scatter, ticks=[0, 1, 2], label='Class')
plt.scatter([c[0] for c in centers], [c[1] for c in centers], 
           marker='*', s=500, c='red', edgecolors='white', linewidth=2, 
           label='Neuron Centers')

plt.xlabel(feature_names_2d[0])
plt.ylabel(feature_names_2d[1])
plt.title('RBF Neuron Decision Boundaries (2D Projection)')
plt.legend()
plt.show()
```
