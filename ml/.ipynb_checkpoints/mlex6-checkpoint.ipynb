### AIM

To analyze the impact of dimensionality reduction using Principal Component Analysis (PCA) on the performance and stability of various machine learning classifiers by comparing models trained with and without PCA using 5-fold cross-validation.

### OBJECTIVE

- To preprocess and standardize the dataset before model training.

- To apply PCA for dimensionality reduction while retaining 95% variance.

- To train multiple classifiers in the original feature space (No-PCA).

- To train the same classifiers in the reduced feature space (With-PCA).

- To evaluate performance using accuracy, F1-score, and 5-fold cross-validation.

- To compare stability and overfitting behavior across models.

### Support Vector Machine (SVM), Naive Bayes, k-Nearest Neighbors (KNN)

#### Dataset used
- Spambase dataset
- From assignment-3

```python
import pandas as pd
#LOADING THE DATASET
df=pd.read_csv("spambase_csv.csv")
print("Number of Samples :", df.shape[0])
print("Number of Features :", df.shape[1] - 1) 
```

```python
print(df.columns)
print(df.shape)
```

```python
#STATISTICS
df.describe()
```

```python
#FEATURES
x=df.drop(["class","capital_run_length_average","capital_run_length_longest","capital_run_length_total"],axis=1)
y=df["class"]

#Normalizing or standardizing the features
from sklearn.preprocessing import StandardScaler,MinMaxScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x="class",data=df)
plt.title("Spam Distribution")
plt.show()

sns.scatterplot(data=df,x="word_freq_free",y="word_freq_money",hue="class")
plt.title("FREE WORD vs MONEY WORD")
plt.show
plt.show()
```

```python
#SPLITTING DATA INTO TRAIN SET AND TEST SET
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
models={"SVM":(SVC(probability=True),{"kernel":["linear","rbf"],"C":[0.1,1,10],"gamma":["scale","auto"]}),
        "Naive Bayes":(GaussianNB(),{"var_smoothing":np.logspace(0,-9,num=100)}),
        "KNN":(KNeighborsClassifier(),{"n_neighbors":[3,5,7],"weights":["uniform","distance"],"metric":["euclidean","manhattan"]}),
       }
```

```python
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.metrics import accuracy_score,f1_score

#without pca

def evaluate(name, m, params, x_tr, x_te):
    if params:
        grid = GridSearchCV(m, params, cv=5, scoring="accuracy")
        grid.fit(x_tr, y_train)
        best = grid.best_estimator_
    else:
        best = m.fit(x_tr, y_train)
    y_pred = best.predict(x_te)
    acc = accuracy_score(y_test, y_pred)
    f1  = f1_score(y_test, y_pred, average="weighted")
    cv_scores = cross_val_score(best, x_tr, y_train, cv=5)
    return acc, f1, cv_scores, best
result_no_pca = {}
best_models_no_pca = {}
for name, (m, param) in models.items():
    acc, f1, cv, best = evaluate(name, m, param, x_train, x_test)
    result_no_pca[name] = (acc, f1, cv)
    best_models_no_pca[name] = best
```

```python
print("PERFORMANCE WITHOUT PCA")
df1 = pd.DataFrame.from_dict(result_no_pca, orient='index', columns=['Accuracy', 'F1-Score', 'CV Score'])
folds_df = df1['CV Score'].apply(pd.Series)
folds_df.columns = [f'Fold {i+1}' for i in range(folds_df.shape[1])]
df_final = pd.concat([df1.drop(columns=['CV Score']), folds_df], axis=1)
df_final
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)

n_components = pca.n_components_
explained_variance = pca.explained_variance_ratio_.sum() * 100

print("Chosen Components:", n_components)
print("Explained Variance:", explained_variance)
print("ORIGINAL SHAPE:",x_train.shape)

print("REDUCED SHAPE:",x_train_pca.shape)
```

```python
#Scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Scree Plot")
plt.show()
```

```python
pca_table=pd.DataFrame({"Settings":["With-PCA"],"Chosen Components":[f"{n_components}"],"Explained Variance (%)":[f"{explained_variance}"],"Justification": ["Dimensionality reduced while retaining most data variance and minimizing information loss."]})
pca_table
```

```python
#with pca
result_pca = {}
best_models_pca = {}
for name, (m, param) in models.items():
    acc, f1, cv, best = evaluate(name, m, param,x_train_pca, x_test_pca)
    result_pca[name] = (acc, f1, cv)
    best_models_pca[name] = best
```

```python
print("PERFORMANCE WITHOUT PCA")
df2 = pd.DataFrame.from_dict(result_pca, orient='index', columns=['Accuracy', 'F1-Score', 'CV Score'])
folds_df = df2['CV Score'].apply(pd.Series)
folds_df.columns = [f'Fold {i+1}' for i in range(folds_df.shape[1])]
df_final = pd.concat([df2.drop(columns=['CV Score']), folds_df], axis=1)
df_final
```

```python
print("SVM — Hyperparameter Tuning Results")
svm_table = pd.DataFrame({"Kernel": ["linear / rbf"],"C Values Tried": ["0.1, 1, 10"],"Gamma Values Tried": ["scale, auto"],"Performance (No PCA)": [result_no_pca["SVM"][0]],"Performance (With PCA)": [result_pca["SVM"][0]]})
svm_table
```

```python
print("Naive Bayes — Smoothing Choices")
nb_table = pd.DataFrame({"Smoothing Parameter": ["logspace(0, -9)"],"Performance (No PCA)": [result_no_pca["Naive Bayes"][0]],"Performance (With PCA)": [result_pca["Naive Bayes"][0]]})
nb_table
```

```python
print("KNN — Hyperparameter Tuning")
knn_table = pd.DataFrame({"k Values": ["3, 5, 7"],"Weights": ["uniform, distance"],"Distance Metrics": ["euclidean, manhattan"],"Performance (No PCA)": [result_no_pca["KNN"][0]],"Performance (With PCA)": [result_pca["KNN"][0]]})
knn_table
```

```python
from sklearn.metrics import roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def plot_roc_cm(model, X_test_used, y_test, title):
    y_prob = model.predict_proba(X_test_used)[:, 1]
    # ROC
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
    plt.plot([0,1],[0,1],'--')
    plt.title(f"ROC Curve — {title}")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Confusion Matrix
    y_pred = model.predict(X_test_used)
    cm = confusion_matrix(y_test, y_pred)
    ConfusionMatrixDisplay(cm).plot(cmap="Blues")
    plt.title(f"Confusion Matrix — {title}")
    plt.show()
```

```python
plot_roc_cm(best_models_no_pca["SVM"], x_test, y_test, "SVM (No PCA)")
plot_roc_cm(best_models_no_pca["Naive Bayes"], x_test, y_test, "Naive Bayes (No PCA)")
plot_roc_cm(best_models_no_pca["KNN"], x_test, y_test, "KNN (No PCA)")
```

```python
plot_roc_cm(best_models_pca["SVM"], x_test_pca, y_test, "SVM (With PCA)")
plot_roc_cm(best_models_pca["Naive Bayes"], x_test_pca, y_test, "Naive Bayes (With PCA)")
plot_roc_cm(best_models_pca["KNN"], x_test_pca, y_test, "KNN (With PCA)")
```

```python
#dropping unessential columns
df_log=df_log.drop("Customer ID",axis=1)
df_log=df_log.drop("Name",axis=1)
```

```python
# Handling missing values
for col in df_log.columns:
    if df_log[col].dtype=='object':
        df_log[col]=df_log[col].fillna(df_log[col].mode()[0])
    else:
        df_log[col]=df_log[col].fillna(df_log[col].median())
```

```python
#one hot encoding
df_log=pd.get_dummies(df_log,drop_first=True)
```

```python
#Outlier Handling
df_log = df_log[df_log["Income (USD)"] < df_log["Income (USD)"].quantile(0.99)]
df_log = df_log[df_log["Property Price"] < df_log["Property Price"].quantile(0.99)]

df_log["log_income"] = np.log1p(df_log["Income (USD)"])
median=df_log.loc[df_log["Current Loan Expenses (USD)"] >=0,"Current Loan Expenses (USD)"].median()
df_log.loc[df_log["Current Loan Expenses (USD)"]<0,"Current Loan Expenses (USD)"]=median
df_log["log_expense"] = np.log1p(df_log["Current Loan Expenses (USD)"])
```

```python
df_log['Loan Sanction Amount (USD)']=df_log['Loan Sanction Amount (USD)'].fillna(df_log['Loan Sanction Amount (USD)'].median())
(df_log["Loan Sanction Amount (USD)"]<0).sum()
median=df_log.loc[df_log["Loan Sanction Amount (USD)"] >=0,"Loan Sanction Amount (USD)"].median()
df_log.loc[df_log["Loan Sanction Amount (USD)"]<0,"Loan Sanction Amount (USD)"]=median
df_log['log_loan_amount'] = np.log1p(df_log['Loan Sanction Amount (USD)'])
```

```python
#Addding Ratio-Based Features
df_log["Loan_to_Income"] = df_log["Loan Amount Request (USD)"] / (df_log["Income (USD)"] + 1)
```

```python
#selecting features(x,y)
x=df_log.drop(["Loan Sanction Amount (USD)"],axis=1)
y=df_log["Loan Sanction Amount (USD)"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, StackingClassifier
from xgboost import XGBClassifier
dt_params = {"max_depth": [3,5,10],"min_samples_split": [2,5,10]}
rf_params = {"n_estimators": [50,100],"max_depth": [5,10]}
ada_params = {"n_estimators": [50,100],"learning_rate": [0.5,1]}
gb_params = {"n_estimators": [50,100],"learning_rate": [0.05,0.1]}
xgb_params = {"n_estimators": [50,100],"max_depth": [3,5],"learning_rate": [0.05,0.1]}

def evaluate_model(model, params, x_tr, x_te):
    grid = GridSearchCV(model, params, cv=5, scoring="accuracy")
    grid.fit(x_tr, y_train)
    best = grid.best_estimator_
    y_pred = best.predict(x_te)
    acc = accuracy_score(y_test, y_pred)
    f1  = f1_score(y_test, y_pred, average="weighted")
    cv  = cross_val_score(best, x_tr, y_train, cv=5)
    return acc, f1, cv,best
```

```python
#without pca
dt_no  = evaluate_model(DecisionTreeClassifier(), dt_params, x_train, x_test)
rf_no  = evaluate_model(RandomForestClassifier(), rf_params, x_train, x_test)
ada_no = evaluate_model(AdaBoostClassifier(), ada_params, x_train, x_test)
gb_no  = evaluate_model(GradientBoostingClassifier(), gb_params, x_train, x_test)
xgb_no = evaluate_model(XGBClassifier(eval_metric="logloss"), xgb_params, x_train, x_test)
```

```python
#with pca
dt_pca  = evaluate_model(DecisionTreeClassifier(), dt_params, x_train_pca, x_test_pca)
rf_pca  = evaluate_model(RandomForestClassifier(), rf_params, x_train_pca, x_test_pca)
ada_pca = evaluate_model(AdaBoostClassifier(), ada_params, x_train_pca, x_test_pca)
gb_pca  = evaluate_model(GradientBoostingClassifier(), gb_params, x_train_pca, x_test_pca)
xgb_pca = evaluate_model(XGBClassifier(eval_metric="logloss"), xgb_params, x_train_pca, x_test_pca)
```

### OBSERVATIONS

#### 1. Which models improved most with PCA? Which did not? Why?

**Improved most:**  
Decision Tree, Gradient Boosting, and XGBoost showed improvement in average accuracy with PCA.

**Did not improve / slightly decreased:**  
SVM, Naïve Bayes, KNN, Logistic Regression, Random Forest, and AdaBoost showed slight performance drop.

**Reason:**  
PCA removes feature variance and redundancy. While some ensemble methods benefit from reduced noise, certain models may lose useful discriminative information.

---

#### 2. Did PCA reduce variance across folds (more stable results)?

Yes. PCA slightly reduced fold variation for Decision Tree and boosting models, indicating improved stability.  
For linear models, variance change was minimal.

---

#### 3. For high-dimensional data, was PCA beneficial in reducing overfitting?

Yes. PCA reduced dimensionality and noise, helping models like Decision Tree and boosting methods generalize better, thereby reducing overfitting.

---

#### 4. How did linear models behave compared to ensemble models with PCA?

Linear models (Logistic Regression, SVM) showed slight performance decrease or remained stable.  
Ensemble models, especially boosting methods, showed better adaptability and in some cases improved performance with PCA.

---

#### 5. Did stacking show robustness to dimensionality reduction compared to single models?

Yes. Stacking maintained high performance in both cases, demonstrating robustness to PCA since it combines multiple learners and balances dimensionality effects.

### FINAL CONCLUSION

- PCA was beneficial for high-dimensional data with correlated and redundant features.
- It helped reduce noise and overfitting by retaining only important variance.
- Boosting and some ensemble models showed improvement or stable performance after PCA.
- Linear models like Logistic Regression and SVM showed only slight change in performance.
- Tree-based models such as Random Forest and AdaBoost did not benefit much due to built-in feature selection.
- Stacking remained robust in both cases, showing stability to dimensionality reduction.
- Overall, PCA is helpful when redundancy and high dimensionality exist, but less impactful when models already manage feature selection effectively.
