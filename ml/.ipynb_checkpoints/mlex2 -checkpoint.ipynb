```python
#dropping unessential columns
df=df.drop("Customer ID",axis=1)
df=df.drop("Name",axis=1)
```

```python
# Handling missing values
for col in df.columns:
    if df[col].dtype=='object':
        df[col]=df[col].fillna(df[col].mode()[0])
    else:
        df[col]=df[col].fillna(df[col].median())
```

```python
#one hot encoding
df=pd.get_dummies(df,drop_first=True)
```

```python
#Outlier Handling
df = df[df["Income (USD)"] < df["Income (USD)"].quantile(0.99)]
df = df[df["Property Price"] < df["Property Price"].quantile(0.99)]

df["log_income"] = np.log1p(df["Income (USD)"])
median=df.loc[df["Current Loan Expenses (USD)"] >=0,"Current Loan Expenses (USD)"].median()
df.loc[df["Current Loan Expenses (USD)"]<0,"Current Loan Expenses (USD)"]=median
df["log_expense"] = np.log1p(df["Current Loan Expenses (USD)"])
```

```python
df['Loan Sanction Amount (USD)']=df['Loan Sanction Amount (USD)'].fillna(df['Loan Sanction Amount (USD)'].median())
```

```python
(df["Loan Sanction Amount (USD)"]<0).sum()
```

```python
median=df.loc[df["Loan Sanction Amount (USD)"] >=0,"Loan Sanction Amount (USD)"].median()
df.loc[df["Loan Sanction Amount (USD)"]<0,"Loan Sanction Amount (USD)"]=median
```

```python
df['log_loan_amount'] = np.log1p(df['Loan Sanction Amount (USD)'])
```

```python
#Addding Ratio-Based Features
df["Loan_to_Income"] = df["Loan Amount Request (USD)"] / (df["Income (USD)"] + 1)
```

```python
#selecting features(x,y)
x=df.drop(["Loan Sanction Amount (USD)"],axis=1)
y=df["Loan Sanction Amount (USD)"]
```

```python
#model training
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_scaled,y)
print("Intercept(b)",lr.intercept_)
print("Coefficient(w)",lr.coef_)
```

```python
#splitting data
from sklearn.model_selection import train_test_split
x_train,x_temp,y_train,y_temp=train_test_split(x_scaled,y,test_size=0.3,random_state=42)
x_val,x_test,y_val,y_test=train_test_split(x_temp,y_temp,test_size=0.5,random_state=42)
```

```python
#model training
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#predicting
y_test_pred=model.predict(x_test)
y_val_pred=model.predict(x_val)
```

```python
#model evaluation
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse=mean_squared_error(y_test,y_test_pred)
mae=mean_absolute_error(y_test,y_test_pred)
r2=r2_score(y_test,y_test_pred)
import numpy as np
rmse=np.sqrt(mse)
print("PERFORMANCE")
print("Mean Sqaure Error:",mse)
print("Mean Absolute Error:",mae)
print("R-Mean Sqaure Error:",rmse)
print("R2 SCORE:",r2)
```

```python
# K-Fold Cross-Validation
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,shuffle=True,random_state=42)
fold=1
mae_list,mse_list,rmse_list,r2_list=[],[],[],[]

for train,test in kf.split(x_scaled):
    train_x,test_x=x_scaled[train],x_scaled[test]
    train_y,test_y=y.iloc[train],y.iloc[test]

    lr=LinearRegression()
    lr.fit(train_x,train_y)
    
    y_pred=lr.predict(test_x)
    
    mae = mean_absolute_error(test_y, y_pred)
    mse = mean_squared_error(test_y, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(test_y, y_pred)
    print(f"Fold {fold} : MAE:{mae:.2f}, MSE:{mse:.2f}, RMSE:{rmse:.2f}, R2:{r2:.2f}")
    
    mae_list.append(mae)
    mse_list.append(mse)
    rmse_list.append(rmse)
    r2_list.append(r2)
    
    fold += 1
    
print("\nAVERAGE CROSS VALIDATION METRICS ")
print("Average MAE :", np.mean(mae_list))
print("Average MSE :", np.mean(mse_list))
print("Average RMSE:", np.mean(rmse_list))
print("Average R2 :", np.mean(r2_list))
```

```python
#SPLITTING INTO DIFFERENT RATIOS
splits = [(0.7,0.3), (0.6,0.4), (0.75,0.25), (0.80,0.20)]

for tr_ratio, te_ratio in splits:
    X_train, X_test, y_train, y_test = train_test_split(
        x_scaled, y, test_size=te_ratio, random_state=42
    )

    lr = LinearRegression()
    lr.fit(X_train, y_train)

    print("SPLIT:", int(tr_ratio*100), "/", int(te_ratio*100))
    print("Train size:", len(y_train))
    print("Test size :", len(y_test))
    r2 = r2_score(y_test, lr.predict(X_test))
    N = X_test.shape[0] + X_train.shape[0]
    p = X_train.shape[1]
    adjusted_r2 = 1 - (1 - r2) * (N - 1) / (N - p - 1)
    print("R2:", r2)
    print("Adjusted R2:", adjusted_r2)
    print("MAE:", mean_absolute_error(y_test, lr.predict(X_test)))
    print("MSE:", mean_squared_error(y_test, lr.predict(X_test)))
    print("----------------------")
```
